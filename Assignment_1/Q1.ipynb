```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
emma=cv.imread('a1images/a1images/emma.jpg', cv.IMREAD_GRAYSCALE)
```

```python
#Defining the transformations for the 3 segments
t1=np.linspace(0,50,50-0+1).astype('uint8')
t2=np.linspace(100,255,150-50).astype('uint8')
t3=np.linspace(150,255,255-150).astype('uint8')

#Concatenate all segments to create the complete transformation
transform = np.concatenate((t1, t2, t3)).astype('uint8')
print(len(transform))

#plotting the transformation curve
plt.plot(np.arange(256), transform)  
plt.title("Intensity Mapping Curve")
plt.xlabel("Input pixel value")
plt.ylabel("Output pixel value")
plt.grid(True)
plt.show()
```

```python
#Applyijng the transformation to the image
image_transformed = cv.LUT(emma, transform)

# Plot original and transformed image
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.title("Original Image")
plt.imshow(emma, cmap='gray')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.title("Transformed Image")
plt.imshow(image_transformed, cmap='gray')
plt.axis('off')

plt.show()
```
