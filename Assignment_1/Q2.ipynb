### (a) white matter

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
brain = cv.imread('a1images/a1images/brain_proton_density_slice.png', cv.IMREAD_GRAYSCALE)
# Transformation curve for white matter 
mu = 200
left = 170   # start of triangle base
right = 230  # end of triangle base

t1 = np.linspace(0, 255, mu - left + 1).astype('uint8')
t2 = np.linspace(255, 0, right - mu + 1).astype('uint8')

# Combine into one LUT
t = np.concatenate((
    np.zeros(left, dtype='uint8'),  # before triangle starts
    t1[1:],                         # rising part
    t2[1:],                         # falling part
    np.zeros(255 - right + 1, dtype='uint8')  # after triangle ends
))

# Plot
x = np.arange(256)
plt.figure(figsize=(5, 5))
plt.plot(x, t)
plt.xlabel("Input intensity")
plt.ylabel("Output intensity")
plt.grid(True)
plt.title(f"Triangular Transform centered at mu={mu}")
plt.show()
```

```python
g=t[brain]

# Display the image
plt.figure(figsize=(5, 5))
plt.imshow(g, cmap='gray', vmin=0, vmax=255)
plt.axis('off')
plt.show()
```

### (b) Gray matter

```python
brain = cv.imread('a1images/a1images/brain_proton_density_slice.png', cv.IMREAD_GRAYSCALE)

#Transformation curve for grey matter

# Parameters
mu = 150
left = 112   # start of triangle base
right = 188  # end of triangle base

t1 = np.linspace(0, 255, mu - left + 1).astype('uint8')
t2 = np.linspace(255, 0, right - mu + 1).astype('uint8')

# Combine into one lookup table
t = np.concatenate((
    np.zeros(left, dtype='uint8'),   # before triangle starts
    t1[1:],                          # rising part
    t2[1:],                          # falling part
    np.zeros(255 - right + 1, dtype='uint8')  # after triangle ends
))

# Plot
x = np.arange(256)
plt.figure(figsize=(5, 5))
plt.plot(x, t)
plt.xlabel("Input intensity")
plt.ylabel("Output intensity")
plt.xlim(0, 280)
plt.ylim(0, 280)
plt.grid(True)
plt.title(f"Triangular Transform centered at mu={mu}")
plt.show()
```

```python
g=t[brain]

# Display the image
plt.figure(figsize=(5, 5))
plt.imshow(g, cmap='gray', vmin=0, vmax=255)
plt.axis('off')
plt.show()
```
