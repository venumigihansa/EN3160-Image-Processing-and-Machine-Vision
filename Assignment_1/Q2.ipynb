### (a) white matter

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
brain = cv.imread('a1images/a1images/brain_proton_density_slice.png', cv.IMREAD_GRAYSCALE)

#Transformation curve for white matter
mu=200
sigma=10
x=np.linspace(0, 255, 256)
t=255*np.exp(-((x-mu)**2)/(2*sigma)**2)
t=np.clip(t,0,255)


plt.figure(figsize=(5, 5))
plt.plot(x, t)
plt.xlabel("Input intensity")
plt.xlim(0, 280)
plt.ylim(0, 280)
plt.ylabel("Output intensity")
plt.grid(True)
plt.title(f"Gaussian Transform shifted to mu={mu}")
plt.show()
```

```python
g=t[brain]

# Display the image
plt.figure(figsize=(5, 5))
plt.imshow(g, cmap='gray', vmin=0, vmax=255)
plt.axis('off')
plt.show()
```
