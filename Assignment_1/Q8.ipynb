```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
def zoom(img, technique, scale=4):
    if technique == 'nn':
        return cv.resize(img, None, fx=scale, fy=scale, interpolation=cv.INTER_NEAREST)
    elif technique == 'bilinear':
        return cv.resize(img, None, fx=scale, fy=scale, interpolation=cv.INTER_LINEAR)

def norm_SSD(img1, img2):
    return np.sum((img1 - img2)**2) / img1.size
```

```python
im1 = cv.imread('a1images/a1images/a1q5images/im01.png')
im1_small = cv.imread('a1images/a1images/a1q5images/im01small.png')

im1_zoomed_nn = zoom(im1_small, 'nn')
im1_zoomed_bilinear = zoom(im1_small, 'bilinear')

nn_ssd = norm_SSD(im1, im1_zoomed_nn)
bilinear_ssd = norm_SSD(im1, im1_zoomed_bilinear)

print(f'Image 1 - NN SSD: {nn_ssd:.6f}')
print(f'Image 1 - Bilinear SSD: {bilinear_ssd:.6f}')
```

```python
fig, axs = plt.subplots(1, 2, figsize=(10, 5))
axs[0].imshow(cv.cvtColor(im1_zoomed_nn, cv.COLOR_BGR2RGB))
axs[0].set_title(f'NN (SSD: {nn_ssd:.5f})')
axs[1].imshow(cv.cvtColor(im1_zoomed_bilinear, cv.COLOR_BGR2RGB))
axs[1].set_title(f'Bilinear (SSD: {bilinear_ssd:.5f})')
plt.show()
```

```python
im2 = cv.imread('a1images/a1images/a1q5images/im02.png')
im2_small = cv.imread('a1images/a1images/a1q5images/im02small.png')

im2_zoomed_nn = zoom(im2_small, 'nn')
im2_zoomed_bilinear = zoom(im2_small, 'bilinear')

nn_ssd = norm_SSD(im2, im2_zoomed_nn)
bilinear_ssd = norm_SSD(im2, im2_zoomed_bilinear)

print(f'Image 2 - NN SSD: {nn_ssd:.6f}')
print(f'Image 2 - Bilinear SSD: {bilinear_ssd:.6f}')

fig, axs = plt.subplots(1, 2, figsize=(10, 5))
axs[0].imshow(cv.cvtColor(im2_zoomed_nn, cv.COLOR_BGR2RGB))
axs[0].set_title(f'NN (SSD: {nn_ssd:.5f})')
axs[1].imshow(cv.cvtColor(im2_zoomed_bilinear, cv.COLOR_BGR2RGB))
axs[1].set_title(f'Bilinear (SSD: {bilinear_ssd:.5f})')
plt.show()
```
